```python
# Desempeño del modelo de regresion

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular el error cuadrático medio
mse = mean_squared_error(Y_test, predicciones)
print("Mean Squared Error:", mse)

# Calcular el error absoluto medio
mae = mean_absolute_error(Y_test, predicciones)
print("Mean Absolute Error:", mae)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(Y_test, predicciones)
print("R^2 Score:", r2)
```

```python
# Matrix de confusion de regresion

import matplotlib.pyplot as plt
import seaborn as sns

# Calcular los residuos
residuos = Y_test - Prediccion

# Histograma de residuos
plt.figure(figsize=(8, 6))
sns.histplot(residuos, kde=True, color='blue', bins=30)
plt.title('Histograma de Residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de dispersión de valores reales vs. predichos
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, Prediccion, color='red', alpha=0.5)
plt.title('Valores Reales vs. Predichos')
plt.xlabel('Valores Reales')
plt.ylabel('Valores Predichos')
plt.show()

# Gráfico de línea de regresión
plt.figure(figsize=(8, 6))
plt.plot(Y_test, color='blue', label='Valores Reales')
plt.plot(Prediccion, color='red', linestyle='dashed', label='Valores Predichos')
plt.title('Gráfico de Línea de Regresión')
plt.xlabel('Índice')
plt.ylabel('Valores')
plt.legend()
plt.show()
```
